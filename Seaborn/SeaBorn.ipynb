```python
import seaborn as sns
```

```python

import matplotlib.pyplot as plt
```

```python
# Part 1 Distribution plots
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
#kernel density estimation
#sns.distplot(tips['total_bill'])
#sns.distplot(tips['total_bill'],kde=False)
sns.distplot(tips['total_bill'],kde=False,bins=40) # bin to get more detial 
```

```python
#kind = hex,reg,kde,default is scatter
sns.jointplot(x='total_bill',y='tip',data=tips)
```

```python
sns.pairplot(tips,hue='sex',palette='coolwarm')
```

```python
sns.rugplot(tips['total_bill'])
```

```python
sns.distplot(tips['total_bill'])
```

```python
sns.kdeplot(tips['total_bill'])
```

```python
import numpy as np
```

```python
sns.barplot(x='sex',y='total_bill',data=tips,estimator=np.std)
```

```python
sns.countplot(x='sex',data=tips)
```

```python
#sns.boxplot(x='day',y='total_bill',data=tips,hue='smoker')
sns.boxplot(x='day',y='total_bill',data=tips)
```

```python
sns.violinplot(x='day',y='total_bill',data=tips,hue='sex',split=True)
```

```python
sns.stripplot(x='day',y='total_bill',data=tips,jitter=True,hue='sex',split=True)
```

```python
sns.violinplot(x='day',y='total_bill',data=tips)
sns.swarmplot(x='day',y='total_bill',data=tips,color='black')# not use for large dataset
# if u like mix it wit violinplot
```

```python
sns.factorplot(x='day',y='total_bill',data=tips,kind='bar')# if u want to show violin just put violin in kind
```

```python

# Part 4
# Regression Plots
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex',markers=['o','v'], 
          scatter_kws={'s':100})
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,col='day',row='time',hue='sex',
          aspect=0.6,size=8)
```
